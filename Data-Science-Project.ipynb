```python
def Is_Float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[df3['total_sqft'].apply(Is_Float)]  # This will only have the values in the dataframe that have valid values
```

```python
def convert_sqft_to_num(x):
    var = x.split('-')
    if(len(var) == 2):
        return (float(var[0])+ float(var[1]) / 2)
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_to_num('234-456')
```

```python
convert_sqft_to_num('345=56')
```

```python
df3['total_sqft'] = df3['total_sqft'].apply(convert_sqft_to_num)
```
